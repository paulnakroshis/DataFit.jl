```julia
using Statistics, Distributions
# analyze spacing bewteen elements of a Vector
function spacing_info(v::Vector, full_stats::Bool=false)
    Δt =v[2:end] - v[1:end-1]
    if any(Δt .<= 0)
        error("Vector must be strictly increasing")
    end
    Δtmin = minimum(Δt)
    Δtmax = maximum(Δt)
    Δtmean = mean(Δt)
    Δtstd = std(Δt)
    if full_stats
        Δtmin = minimum(Δt)
        Δtmax = maximum(Δt)
        Δtmean = mean(Δt)
        Δtstd = std(Δt)
        return Δtmin, Δtmax, Δtmean, Δtstd
    else
        return Δtmin
    end
end
```

```julia

t = [1.0, 2.0, 3.0, 3.5, 4.2, 5.0, 6.1, 7.4, 7.6];
tmin =spacing_info(t);
println("tmin = ", tmin)
Δtmin, Δtmax, Δtmean, Δtstd = spacing_info(t, true);
println("Δtmin, Δtmax, Δtmean, Δtstd = ", Δtmin, ", ", Δtmax, ", ", Δtmean, ", ", Δtstd)
```
